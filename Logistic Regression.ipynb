```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
x = np.arange(10).reshape(-1, 1)
y = np.array([0, 1, 1, 0, 1, 1, 1, 1, 1, 1])
print(x)
print(y)
```

```python
model = LogisticRegression(solver='liblinear', C=10.0, random_state=0)
print(model.fit(x, y))
```

```python
p_pred = model.predict_proba(x)
y_pred = model.predict(x)
score_ = model.score(x, y)
conf_m = confusion_matrix(y, y_pred)
report = classification_report(y, y_pred)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("D:\Data for ML\diabetes.csv")
```

```python
X=df.iloc[:,:8].values
print(X)
```

```python
Y=df["Outcome"].values
X_train, X_test, y_train, y_test=train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
model = LogisticRegression(solver='liblinear', C=9.0, random_state=100)
print(model.fit(X_train, y_train))
```

```python
P_pred = model.predict_proba(X_train)
Y_pred = model.predict(X_train)
score_ = model.score(X_train, y_train)
conf_m = confusion_matrix(y_train, Y_pred)
report = classification_report(y_train, Y_pred)
```
