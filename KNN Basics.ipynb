```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df=pd.read_csv("D:\Data for ML\diabetes.csv")
```

```python
X = df.iloc[:,:8].values
print(X)
```

```python
Y=df["Outcome"]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, Y, test_size=0.3, random_state=100)
```

```python
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
P_pred = knn.predict_proba(X_train)
Y_pred = knn.predict(X_train)
score_ = knn.score(X_train, y_train)
conf_m = confusion_matrix(y_train, Y_pred)
```

```python
print(score_)
```

```python
print(conf_m)
```

```python
y_pred = knn.predict(X_test)
count_misclassified = (y_test != y_pred).sum()
print('Misclassified samples: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: {:.2f}'.format(accuracy))
```

```python
y_pred1 = knn.predict(X_train)
count_misclassified = (y_train != y_pred1).sum()
print('Misclassified samples: {}'.format(count_misclassified))
accuracy = metrics.accuracy_score(y_train, y_pred1)
print('Accuracy: {:.2f}'.format(accuracy))
```
